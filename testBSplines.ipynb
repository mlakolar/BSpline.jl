```julia
M = 3
knots = BSpline.create_knots(M,(1,6),6)
num_basis = BSpline.num_basis(M, knots)
```

```julia
X = [-1:0.01:8]
psiX = BSpline.bspline_basismatrix(M, knots, X)
PyPlot.plot(X, psiX)
```

Plot Derivatives

```julia
X = [-1:0.01:8]
psiX = BSpline.derivative_bspline_basismatrix(M, knots, X)
PyPlot.plot(X, psiX)
```
