## Plot BSpline and derivatives

```julia
import BSpline
```

Plot spline functions

```julia
import PyPlot
```

```julia
M = 3
knots = BSpline.create_knots(M,(1,6),6)
num_basis = BSpline.num_basis(M, knots);
```

```julia
X = [-1:0.01:8]
psiX = zeros(length(X), num_basis)
BSpline.bspline_basismatrix!(psiX, M, knots, X)
PyPlot.plot(X, psiX);
```

Plot Derivatives

```julia
X = [-1:0.01:8]
BSpline.derivative_bspline_basismatrix(psiX, M, knots, X)
PyPlot.plot(X, psiX);
```
